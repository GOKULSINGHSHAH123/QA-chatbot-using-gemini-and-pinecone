Approach to Solve the problem

1.Data Splitting:
Start by splitting large business data or big data into manageable chunks. For instance, if you're working with PDFs, you can split them into smaller segments or chunks to handle them more effectively.

2.Vectorization:
Convert these chunks into vectors using Hugging Face embeddings. This step transforms the text chunks into numerical representations that capture their semantic meaning.

3.Storage:
Store these vectors using Pinecone. Pinecone efficiently manages vector storage and enables fast similarity search, handling semantic relationships between context and queries.


Model Interaction:
Utilize the Gemini model (or another suitable model) to generate outputs based on the stored vectors and user queries. Gemini processes the queries and retrieves relevant information from the vectors.
Interface and Testing:

Implement a user interface using Gradio. Gradio allows you to build interactive interfaces to test and interact with your model, making it easier to evaluate performance and usability.

Splitting the whole data to chunks

```python
def create_chunck(doc_to_chunck):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size = 1000,
        chunk_overlap = 200,
        length_function = len)
    chunks = text_splitter.split_documents(doc_to_chunck)
    return chunks

def load_path(path):
    loader = PyPDFLoader(path)
    doc = loader.load()
    return doc

def load_chunck_file(path):
    doc = load_path(path)
    chunks = create_chunck(doc)
    return chunks


file_path = '/content/NIPS-2017-attention-is-all-you-need-Paper.pdf'# Use your path file of any pdf i have used self attention pdf and for business q a bot only small changes will be there in prompt

chunks = load_chunck_file(file_path)
print('N° Chunks:', len(chunks))
print(chunks)
```

```python
from pydantic import ValidationError

# Defining the prompt template
prompt_template = """
Please answer the question with as much detail as possible based on the provided context.
Make sure to include all relevant details. If the answer is not available in the provided context, please respond with
"The answer is not available in the context." Avoid providing incorrect answers.
\n\n
  Context:\n {context}?\n
  Question: \n{question}\n

  Answer:
"""

prompt = PromptTemplate(template=prompt_template, input_variables=["context", "question"])

# Defining the LLM Pinecone index and LLM model
vstore = PineconeVectorStore.from_existing_index(index_name, embeddings)

try:
    llm = ChatGoogleGenerativeAI(model="gemini-pro", google_api_key="AIzaSyAxaNMQlai9ybEiAVlSIOaY-pRivTMuR54", convert_system_message_to_human=True, temperature=0.3)
    chain = load_qa_chain(llm, chain_type="stuff", prompt=prompt)
except ValidationError as e:
    print(e.json())
```

Now deploying to gradio

```python
import gradio as gr

def data_querying(question):
    vstore = PineconeVectorStore.from_existing_index(index_name, embeddings)
    llm = ChatGoogleGenerativeAI(model="gemini-pro", google_api_key="AIzaSyAxaNMQlai9ybEiAVlSIOaY-pRivTMuR54", convert_system_message_to_human=True, temperature=0.3)
    chain = load_qa_chain(llm, chain_type="stuff", prompt=prompt)
    docs = vstore.similarity_search(question, 3)

    response = chain({"input_documents":docs, "question": question}, return_only_outputs=True)
    response_text = response.get('output_text')

    return response_text

data_querying("What is self attention")

iface = gr.Interface(fn=data_querying,
                     inputs=gr.components.Textbox(lines=7, label="Enter your text"),
                     outputs=gr.components.Text(),
                     title="Test DNU Gemini-Pro")

iface.launch()
```
